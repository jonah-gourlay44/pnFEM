```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as io
```

```python
eps_0 = 8.8542e-12 #x10^-13 C/(V*cm)
eps_r_1 = 1
eps_r_2 = 2
rho_1 = 1e-7 #x10^-13 C/(cm^3)
rho_2 = 0
phi_1 = 1.5
phi_2 = 0
L = 0.01
```

```python
num_nodes = 121
num_elements = num_nodes - 1

nodes = np.linspace(0, 2*L, 121)
elements = np.linspace(0, num_nodes-1, num_nodes, dtype=int).repeat(2, axis=0)[1:-1].reshape((-1,2))
x_avg = np.mean(nodes[elements], axis=1).reshape(120,1)

dL = nodes[1] - nodes[0]
```

```python
def eps(x):
    
    eps_func = np.vectorize(lambda x : (eps_r_1 if x < L else eps_r_2) * eps_0)
    
    return eps_func(x)

def rho(x):
    
    rho_func = np.vectorize(lambda x : rho_1 if x < L else rho_2)
    
    return rho_func(x)

def dNi_dNj():
    a = np.zeros((2,2))
    # Why are these not 1/dL instead of dL?
    a[0] = 1 / dL * np.array([1,-1])
    a[1] = 1 / dL * np.array([-1,1])
    
    return a
    
def Ni():
    return dL / 2 * np.array([1,1]) # Why is this not dL instead of dL^2?
```

```python
A = np.zeros((num_nodes, num_nodes))
b = np.zeros((num_nodes,1))

for i in range(0, num_elements):
    nds = elements[i]
    
    eps_x = eps(x_avg[i])#eps(nodes[nds[np.newaxis,:].repeat(2,axis=0)])
    rho_x = rho(x_avg[i])#nodes[nds])
    
    a = dNi_dNj() * eps_x
    b0 = Ni() * rho_x
    
    A[np.ix_(nds,nds)] += a
    b[nds] += b0.reshape((2,1))

A[0][0] = 1; A[0][1:] = 0
A[-1][-1] = 1; A[-1][:-1] = 0
b[0] = phi_1; b[-1] = phi_2
```

```python
x = np.linalg.solve(A,b)
```

```python
plt.figure(figsize=(15,7))
plt.plot(nodes,x)
plt.xlim([0, 0.025])
plt.ylim([0, 1.5])
plt.show()
```
